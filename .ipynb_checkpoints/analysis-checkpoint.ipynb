```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
import sklearn.preprocessing as sp
import seaborn as sns
%matplotlib inline

mp.rcParams['font.family'] = ['sans-serif']
mp.rcParams['font.sans-serif'] = ['SimHei']
mp.rcParams['axes.unicode_minus']=False

cc = pd.read_csv('CC GENERAL.csv',header='infer')
df_cc = pd.DataFrame(cc)
df_cc = df_cc.iloc[:,1:]
print(df_cc.shape)
df_cc.head(10)
```

```python
fig,axes = mp.subplots(nrows=6,ncols=3,figsize=(15,18))
r = 0
c = 0
for item in clo_list:
    if c<=2:
        sns.distplot(df_cc[item],ax=axes[r,c])
        c += 1
    else:
        r += 1
        c = 0
        sns.distplot(df_cc[item],ax=axes[r,c])
        c += 1
```

```python
fig,axes = mp.subplots(nrows=6,ncols=3,figsize=(15,18))
r = 0
c = 0
for item in clo_list:
    if c<=2:
        sns.boxplot(x=item,data=df_cc,ax=axes[r,c])
        c += 1
    else:
        r += 1
        c = 0
        sns.boxplot(x=item,data=df_cc,ax=axes[r,c])
        c += 1
```

```python
for item in clo_list:
    print('************************************************',item,'************************************************')
    q1 = df_cc[item].quantile(q=0.25)
    q2 = df_cc[item].quantile(q=0.5)
    q3 = df_cc[item].quantile(q=0.75)
    q4 = df_cc[item].quantile(q=1)
    print('1% quantile:',q1)
    print('2% quantile:',q2)
    print('3% quantile:',q3)
    print('4% quantile:',q4)
    outliers = df_cc[item].quantile(q=0.75)+df_cc[item].quantile(q=0.75)*1.5-df_cc[item].quantile(q=0.25)
    print('Above:',outliers,'are outliers')
    print('number of outliers:',df_cc[df_cc[item]>outliers][item].count())
    print('percentage:',round(df_cc[df_cc[item]>outliers][item].count()*100/len(df_cc),2),'%')
    print('mean:',round(df_cc[item].mean(),2))
    print('std:',round(df_cc[item].std(),2))
    print('cv:',round(df_cc[item].mean()/df_cc[item].std(),2))
```

```python
import sklearn.cluster as sc
def kmeans_labels(attribute):
    kmodel = sc.KMeans(n_clusters=4,n_jobs=4)
    bp_copy = df_cc[attribute]
    kmodel.fit(bp_copy.values.reshape(len(bp_copy),1))
    centers = pd.DataFrame(kmodel.cluster_centers_).sort_values(by=0)
    w = centers.rolling(2).mean().iloc[1:]
    w = [0]+list(w[0])+[bp_copy.max()]
    df_cc[attribute] = pd.cut(bp_copy,w,labels=range(4),include_lowest=True)

for item in clo_list[0:-1:1]:
    kmeans_labels(item)
```

```python
for item in clo_list:
    print(item,':\n',df_cc[item].isnull().sum())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_cc)
pca_result = pd.DataFrame(pca_result)
pca_result.columns=['p1','p2']
pca_result
```

```python
sse = []
for k in range(1,10):
    km = sc.KMeans(n_clusters=k,init='k-means++',n_init=10,max_iter=300,random_state=10)
    km.fit(pca_result)
    sse.append(km.inertia_)
mp.figure()
mp.title('SSE',fontsize=16)
mp.plot(range(1,10),sse,marker='o')
mp.xlabel('number of clusters',fontsize=14)
mp.ylabel('distortions',fontsize=14)
mp.show()
sse_p = sns.lineplot(x=range(1,10),y=sse)
sse_p.set_title('SSE',fontsize=16)
```

```python
import sklearn.metrics as sm
sil = []
for i in range(2,10):
    v = sm.silhouette_score(pca_result,sc.KMeans(n_clusters=i).fit_predict(pca_result))
    sil.append(v)
```

```python
print(sil)
sil_score = pd.DataFrame(np.arange(2,10,1),columns=['number of clusters'])
y = pd.DataFrame(sil)
sil_p = sns.barplot(x=sil_score['number of clusters'],y=sil)
sil_p.set_title('Silhouette Scores',fontsize=16)
```

```python
kmeans = sc.KMeans(n_clusters=2)
kmeans.fit(pca_result)
mp.figure()
mp.title('K-means Clusters',fontsize=16)
mp.scatter(pca_result['p1'],pca_result['p2'],c=sc.KMeans(n_clusters=2).fit_predict(pca_result),cmap='jet')
mp.show()
```

```python
n = 0.01

l, r = pca_result['p1'].min() - 1, pca_result['p1'].max() + 1
b, t = pca_result['p2'].min() - 1, pca_result['p2'].max() + 1

xx,yy= np.meshgrid(np.arange(l,r,n),np.arange(b,t,n))
z = kmeans.predict(np.array(list(zip(xx.ravel(),yy.ravel()))))
# print(xx.ravel().shape)
# print(yy.ravel().shape)
# print(z.shape)
z =z.reshape(xx.shape)

mp.figure()
mp.clf()
mp.title('K-means clustering on the credit card fraud dataset (PCA-reduced data)\n Centroids are marked with white circle', fontsize=16)
mp.imshow(z,interpolation='nearest',extent=(xx.min(),xx.max(),yy.min(),yy.max()),cmap=mp.cm.winter,aspect='auto',origin='lower')
mp.plot(pca_result['p1'],pca_result['p2'],'k.',markersize=2)
centroids = kmeans.cluster_centers_
mp.scatter(centroids[:,0],centroids[:,1],marker='o',s=10,linewidths=3,color='white',zorder=5)
mp.xlim(l,r)
mp.ylim(b,t)
mp.show()
```
